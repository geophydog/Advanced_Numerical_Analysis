## $\color{red}{\bf{y=Ax+B}}$
$$
\large{
\begin{aligned}
    A &= \frac{N\sum_{k=1}^Nx_ky_k-\sum_{k=1}^Nx_k\sum_{k=1}^Ny_k}{N\sum_{k=1}^Nx_k^2-\sum_{k=1}^Nx_k\sum_{k=1}^Nx_k}\\
    B &= \frac{\sum_{k=1}^Ny_k-A\sum_{k=1}^Nx_k}{N}
\end{aligned}
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def line_fit(x0, y0):
    xy = sum(x0*y0); x = sum(x0); x2 = sum(x0**2)
    y = sum(y0); N = len(x0); A = (N*xy-x*y)/(N*x2-x**2)  
    B = (y-A*x) / N
    return A, B

x0 = np.linspace(0, 10, 21)
y0 = 5*x0 + 3 + np.random.randn(len(x0))
x = np.linspace(0, 10, 101)
A, B = line_fit(x0, y0)
print('A:', A, 'B:', B)
y1 = 5*x +3
y2 = A*x + B
plt.plot(x, y1, lw=3, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted', zorder=1)
plt.plot(x0, y0, 'go', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{y=Ax^M}$
$$
\large{
    A = \frac{\sum_{k=1}^Nx_k^My_k}{\sum_{k=1}^Nx_k^{2M}}
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def power_fit(x0, y0, M):
    xy = sum(x0**M*y0); x2 = sum(x0**(2*M))
    A = xy / x2
    return A

x1 = 3
x0 = np.linspace(0, x1, 21)
M = 3
y0 = 2*x0**M + np.random.randn(len(x0))*2
x = np.linspace(0, x1, 101)
A = power_fit(x0, y0, M)
print('A:', A)
y1 = 2*x**M
y2 = A*x**M
plt.plot(x, y1, lw=5, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{y=Ce^{Ax}}$
$$
\large{
\begin{aligned}
    X_k &= x_k\\
    Y_k &= lny_k\\
    B &= lnC\\
    A &= \frac{N\sum_{k=1}^NX_kY_k-\sum_{k=1}^NX_k\sum_{k=1}^NY_k}{N\sum_{k=1}^NX_k^2-\sum_{k=1}^NX_k\sum_{k=1}^NX_k}\\
    B &= \frac{\sum_{k=1}^NY_k-A\sum_{k=1}^NX_k}{N}
\end{aligned}
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def line_exp_fit(x0, y0):
    y0 = np.log(y0)
    xy = sum(x0*y0); x = sum(x0); x2 = sum(x0**2)
    y = sum(y0); N = len(x0); A = (N*xy-x*y)/(N*x2-x**2)  
    B = (y-A*x) / N; C = np.exp(B)
    return A, C

x1 = 1
x2 = 2
x0 = np.linspace(x1, x2, 11)
y0 = 2*np.exp(2*x0) + np.random.randn(len(x0))*2
x = np.linspace(x1, x2, 101)
A, C = line_exp_fit(x0, y0)
print('A:', A, 'C:', C)
y1 = 2 * np.exp(2*x)
y2 = C * np.exp(A*x)
plt.plot(x, y1, lw=5, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{\bf{y=\frac{D}{x+C}}}$
$$
\large{
\left \{
\begin{aligned}
    X &= xy\\
    Y &= y\\
    C &= \frac{-1}{A}\\
    D &= \frac{-B}{A}\\
    A &= \frac{N\sum_{k=1}^NX_kY_k-\sum_{k=1}^NX_k\sum_{k=1}^NY_k}{N\sum_{k=1}^NX_k^2-\sum_{k=1}^NX_k\sum_{k=1}^NX_k}\\
    B &= \frac{\sum_{k=1}^NY_k-A\sum_{k=1}^NX_k}{N}
\end{aligned}
\right.
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def line_frac_fit(x0, y0):
    x0 = x0*y0
    xy = sum(x0*y0); x = sum(x0); x2 = sum(x0**2)
    y = sum(y0); N = len(x0); A = (N*xy-x*y)/(N*x2-x**2)  
    B = (y-A*x) / N; C = -1 / A; D = -B / A
    return C, D

x1 = 0
x2 = 3
x0 = np.linspace(x1, x2, 21)
y0 = 1 / (x0+2) + np.random.randn(len(x0))*np.arange(len(x0))*1e-3
x = np.linspace(x1, x2, 101)
C, D = line_frac_fit(x0, y0)
print('C:', C, 'D:', D)
y1 = 1 / (x+2)
y2 = D / (x+C)
plt.plot(x, y1, lw=5, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{y=Ax^2+Bx+C}$


```python
import numpy as np
import matplotlib.pyplot as plt

def pow_2_poly_fit(x0, y0):
    y = np.empty(3); m = np.empty((3, 3))
    for i in range(3):
        y[i] = sum(x0**(3-i-1)*y0)
        for j in range(3):
            m[i, j] = sum(x0**(4-i-j))
    return np.linalg.solve(m, y)

x1 = -2; x2 = 2
x0 = np.linspace(x1, x2, 21)
x = np.linspace(x1, x2, 101)
y0 = 2*x0**2 + 3*x0 + 4 +np.random.randn(len(x0))
y1 = 2*x**2 + 3*x + 4
a = pow_2_poly_fit(x0, y0)
y2 = a[0]*x**2 +a[1]*x + a[2]

plt.plot(x, y1, lw=5, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{y=a_0+a_1x+a_2x^2+\cdots+a_Nx^N}$

```python
import numpy as np
import matplotlib.pyplot as plt

def n_power_poly_fit(x0, y0, N=2):
    y = np.empty(N+1)
    m = np.empty((N+1, N+1))
    for i in range(N+1):
        y[i] = sum(x0**(N-i)*y0)
        for j in range(N+1):
            m[i, j] = sum(x0**(2*N-i-j))
    return np.linalg.solve(m, y)

x1 = -10; x2 = 5
x0 = np.linspace(x1, x2, 21)
N = 2
x = np.linspace(x1, x2, 101)
y0 = np.zeros_like(x0)
y1 = np.zeros_like(x)
for i in range(N+1):
    y0 += 1/(i**6+10) * x0**i
    y1 += 1/(i**6+10) * x**i
y0 += np.random.randn(len(y0))*3e-2
a = n_power_poly_fit(x0, y0, N=N)
y2 = np.ones_like(y1)*a[0]
for i in range(1, len(a)):
    y2 = y2*x + a[i]

plt.plot(x, y1, lw=4, label='Analytical', zorder=0)
plt.plot(x, y2, lw=2, label='Fitted up to $x^%d$'%N, zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{f(x)=\frac{x}{Ax+B}}$

$$
\large{  
\left \{
\begin{aligned}
    X &= \frac{1}{x}\\
    Y &= \frac{1}{y}\\
    B &= \frac{N\sum_{k=1}^NX_kY_k-\sum_{k=1}^NX_k\sum_{k=1}^NY_k}{N\sum_{k=1}^NX_k^2-\sum_{k=1}^NX_k\sum_{k=1}^NX_k}\\
    A &= \frac{\sum_{k=1}^NY_k-A\sum_{k=1}^NX_k}{N}
\end{aligned}
\right.
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def line_frac2_fit(x0, y0,):
    N = len(x0); x0 = 1 / x0; y0 = 1 / y0
    x = sum(x0); y = sum(y0); x2 = sum(x0**2); xy = sum(x0*y0)
    B = (N*xy-x*y) / (N*x2-x**2)
    A = (y-B*x) / N
    return A, B

x1 = 1; x2 = 5
x0 = np.linspace(x1, x2, 21)
x = np.linspace(x1, x2, 101)
y0 = x0 / (2*x0+3) + np.random.randn(len(x0))*1e-2
y1 = x / (2*x+3)
A, B = line_frac2_fit(x0, y0)
y2 = x / (A*x+B)
plt.plot(x, y1, lw=4, label='Analytical', zorder=0)
plt.plot(x, y2, lw=2, label='Fitted', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```

## $\color{red}{f(x)=\frac{x}{Ax+B}}$

$$
\large{  
\left \{
\begin{aligned}
    X &= x\\
    Y &= \frac{x}{y}\\
    A &= \frac{N\sum_{k=1}^NX_kY_k-\sum_{k=1}^NX_k\sum_{k=1}^NY_k}{N\sum_{k=1}^NX_k^2-\sum_{k=1}^NX_k\sum_{k=1}^NX_k}\\
    B &= \frac{\sum_{k=1}^NY_k-A\sum_{k=1}^NX_k}{N}
\end{aligned}
\right.
}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

def line_frac3_fit(x0, y0,):
    N = len(x0); y0 = x0 / y0
    x = sum(x0); y = sum(y0); x2 = sum(x0**2); xy = sum(x0*y0)
    A = (N*xy-x*y) / (N*x2-x**2)
    B = (y-A*x) / N
    return A, B

x1 = 1; x2 = 10
x0 = np.linspace(x1, x2, 21)
x = np.linspace(x1, x2, 101)
y0 = x0 / (2*x0+3) + np.random.randn(len(x0))*1e-2
y1 = x / (2*x+3)
A, B = line_frac2_fit(x0, y0)
A2, B2 = line_frac3_fit(x0, y0)
y2 = x / (A*x+B)
y3 = x / (A2*x+B2)
plt.plot(x, y1, lw=5, label='Analytical', zorder=0)
plt.plot(x, y2, lw=3, label='Fitted 1', zorder=1)
plt.plot(x, y3, lw=1.5, label='Fitted 2', zorder=1)
plt.plot(x0, y0, 'mo', label='Samples', zorder=2)
plt.grid(); plt.legend(fontsize=15); plt.show()
```
